```python
## 1
## Array Creation and Manipulation

import numpy as np

# Membuat array 2D dengan ukuran (5, 5) dan angka random antara 1 dan 20
satu = np.random.randint(1, 21, size=(5, 5))
print("Array sebelum modifikasi:")
print(satu)

# Mengganti semua nilai pada kolom ketiga dengan 1
satu[:, 2] = 1  # : berarti semua baris, 2 adalah indeks kolom ketiga (indeks dimulai dari 0)

print("\nArray setelah modifikasi:")
print(satu)

# Membuat array 2D dengan ukuran (4, 4) dan angka random antara 1 dan 16
dua = np.random.randint(1,17, size=(4,4))
print("\nArray Kedua Sebelum Dimodifikasi:\n",dua)

# Mengganti semua nilai diagonal menjadi 0
np.fill_diagonal(dua,0)
print("\nArray Kedua Setelah Dimodifikasi:\n",dua)

```

```python
## 2
## Array Indexing & Slicing
print("Array:")
duwa = np.arange(1,37).reshape((6,6))
print(duwa)
sub_array = duwa[2:5, 1:4]
print("\n=============================")
print("Sub-array:")
print(sub_array)



duwain = np.random.randint(1,100,size=(5,5))
print(duwain)

print("\n=============================")
border_elements = np.concatenate((duwain[0, :], duwain[-1, :], duwain[1:-1, 0], duwain[1:-1, -1]))
print("Border elements:")
print(border_elements)
```

```python
## 3
## Array Operations

print("Array 1: ")
tiga = np.random.randint(1,11,size=(3,4))
print(tiga)
print("==============================")
print("Array 2: ")
tiga2 = np.random.randint(1,11,size=(3,4))
print(tiga2)
print("\n============================")
print("Additon: ")
addition = tiga + tiga2
print(addition)
print("\n============================")
print("Substraction: ")
substraction = tiga - tiga2
print(substraction)
print("\n============================")
print("Multiplication: ")
multiplication = tiga * tiga2
print(multiplication)
print("\n============================")
print("Divison: ")
division = tiga / tiga2
print(division)
```

```python
## 4
## Statistical Operations

four = np.random.randint(1,11,size=(5,5))
print(four)

print("==================================")
print(f"Rata-rata: {np.mean(four)}")
print(f"Median: {np.mean(four)}")
print(f"Standard Deviasi: {np.std(four)}")
print(f"Varinasa: {np.var(four)}")
print(f"Jumlah: {np.sum(four)}")
```
